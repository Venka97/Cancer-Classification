```python
import pandas as pd
import numpy as np
from IPython.display import display

data = pd.read_csv('data.csv')
display(data.head())
display(data.shape[0])
display(data.shape[1])
```

```python
data.drop(["id","Unnamed: 32"], axis = 1, inplace = True)
display(data.tail())
```

```python
from sklearn.preprocessing import LabelEncoder

labels = data['diagnosis']


lb = LabelEncoder()
labels = lb.fit_transform(labels)
display(labels)
data.drop(['diagnosis'] , axis = 1, inplace = True)
```

```python
display(data.head())
```

```python
from sklearn.preprocessing import StandardScaler

numerical = ['radius_mean','texture_mean','perimeter_mean','area_mean','smoothness_mean','compactness_mean','concavity_mean','concave points_mean','symmetry_mean','fractal_dimension_mean','radius_worst','texture_worst','perimeter_worst','area_worst','smoothness_worst','compactness_worst','concavity_worst','concave points_worst','symmetry_worst','fractal_dimension_worst']
scaler = StandardScaler()
data[numerical] = scaler.fit_transform(data[numerical])
display(data.head())
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components = 10, random_state = 42)
pca.fit_transform(data)

display(pca.explained_variance_ratio_)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test  = train_test_split(data, labels, test_size = 0.2, random_state = 42)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from xgboost.sklearn import XGBClassifier
from sklearn.metrics import make_scorer, accuracy_score, f1_score



f_params = {'objective': 'binary_logistic', 'silent': 1}

clf1 = RandomForestClassifier()
clf2 = LogisticRegression()
clf3 = XGBClassifier(**f_params)

params1 = { 'n_estimators' : [5,10,15,20], 'max_depth' : [1,2,3]}
params2 = {'penalty': ('l1', 'l2'), 'C' : [1,3,5]}
params3 = {'max_depth' : [3,5,10], 'n_estimators': [5,10,15], 'learning_rate' : (1, 3)}


cv = StratifiedKFold(n_splits = 5, shuffle = True, random_state = 42)


clf = GridSearchCV(clf3, cv = cv, param_grid = params3)
clf.fit(X_train,y_train)
preds = clf.predict(X_test)

print "The accuracy is {}%".format(accuracy_score(preds, y_test)*100)
print "The accuracy is {}%".format(f1_score(preds, y_test)*100)
```
